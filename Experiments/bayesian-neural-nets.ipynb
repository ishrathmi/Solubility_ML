### Specific solvents for now

```python
data_subset = data[data["Solvent_1"] == "Methanol"]
data_subset = data_subset[data_subset["Solvent_2"] == "Water"]
```

```python
data_subset.describe()
```

## Train-test split

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(data_subset, test_size=0.2)
```

## Features

### Molecular features

```python
from rdkit.Chem import AllChem
from rdkit import Chem

import torch

def get_mol_feats(smiles_list):
    generator = AllChem.GetMorganGenerator(radius=2, fpSize=1024)
    mols = [Chem.MolFromSmiles(smiles) for smiles in smiles_list]
    return torch.tensor(
        np.asarray([generator.GetFingerprintAsNumPy(mol) for mol in mols])
    ).float()

X_mol_train = get_mol_feats(df_train["Drug_SMILES"])
X_mol_test = get_mol_feats(df_test["Drug_SMILES"])

print(X_mol_train.shape, X_mol_test.shape)
```

### Numerical features (contexts)

```python
from sklearn import preprocessing

NUM_COLS = [
    "Solvent_mass_fraction", 
    "Solvent_mol_fraction", 
    "Temperature (K)", 
    "Drug_Collected_Melting_temp (K)",
    "Drug_Predicted_Melting_temp (K)",
]

scaler = preprocessing.StandardScaler().fit(df_train[NUM_COLS])

def get_num_feats(df):
    return torch.tensor(scaler.transform(df[NUM_COLS])).float()

X_num_train = get_num_feats(df_train)
X_num_test = get_num_feats(df_test)

print(X_num_train.shape, X_num_test.shape)
```

### Combined features

```python
X_train = torch.cat([X_mol_train, X_num_train], dim=-1)
X_test = torch.cat([X_mol_test, X_num_test], dim=-1)

# X_train = X_mol_train
# X_test = X_mol_test

# X_train = X_num_train
# X_test = X_num_test

print(X_train.shape, X_test.shape)
```

### Targets

```python
def get_targets(df):
    return torch.tensor(df["LogS"].to_numpy()).float().unsqueeze(-1)

y_train = get_targets(df_train)
y_test = get_targets(df_test)

print(y_train.shape, y_test.shape)
```

## MLP

```python
from torch import nn


model = nn.Sequential(
    nn.Linear(X_train.shape[-1], 100),
    nn.ReLU(),
    nn.Linear(100, 50),
    nn.ReLU(),
    nn.Linear(50, 1),
)
```

```python
import torch.utils.data as data_utils
from torch import optim

BATCH_SIZE = 128
N_EPOCHS = 500

train_loader = data_utils.DataLoader(
    data_utils.TensorDataset(X_train, y_train),
    shuffle=True,
    batch_size=BATCH_SIZE,
)

opt = optim.AdamW(model.parameters(), lr=1e-3, weight_decay=5e-4)
sched = optim.lr_scheduler.CosineAnnealingLR(opt, N_EPOCHS * len(train_loader))
loss_fn = nn.MSELoss()

test_metric = nn.L1Loss()
```

```python
from tqdm.notebook import trange

pbar = trange(N_EPOCHS)

for epoch in pbar:
    for x, y in train_loader:
        opt.zero_grad()
        model.train()
        
        out = model(x)
        loss = loss_fn(out, y)
        loss.backward()
        
        opt.step()
        sched.step()

    all_out = model(X_train)
    train_mse = loss_fn(all_out, y_train)
    train_mae = test_metric(all_out, y_train)
    pbar.set_description(f"[Train MSE: {train_mse:.4f}, Train MAE: {train_mae:.4f}]")
```

```python
out_test = model(X_test)
test_mse = loss_fn(out_test, y_test)
test_mae = test_metric(out_test, y_test)

print(f"Test MSE: {test_mse:.4f}, Test MAE: {test_mae:.4f}")
```

### Bayesian model

Using the Laplace Approximation, we turn the previous neural network into a Bayesian neural network.

```python
from laplace import Laplace

la = Laplace(
    model, 
    likelihood="regression", 
    subset_of_weights="all", 
    hessian_structure="kron",
)
la.fit(train_loader)
la.optimize_prior_precision()
```

The MSE and MAE on the test set will be the same as before.

```python
out_test, var_test = la(X_test)
test_mse = loss_fn(out_test, y_test)
test_mae = test_metric(out_test, y_test)

print(f"Test MSE: {test_mse:.4f}, Test MAE: {test_mae:.4f}")
```

But now we will have an error-bar (variance of the predicted LogS), which can be seen as our confidence interval around our LogS predictions.
